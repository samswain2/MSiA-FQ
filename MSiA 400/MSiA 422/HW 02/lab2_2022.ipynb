## functions

````python
import random

def genElement (N):
    """
    Input: Length of the list to be generated, N
    Output: List of N numbers and words
    The procedural function to randomly generate a list of N numbers and words.
    """
    # setting the seed and counter for type
    #rng.current = N
    
    L = []

    word_count = 0
    num_count = 0

    while (word_count + num_count) < N:

    #for x in range(N):
        #randomizer for number or word
        NorW = random.random()
        if NorW < 0.5:
            Type = "number"          
        else:
            Type = 'word'
            
        #generating random number from 0-100 and add to the list
        if Type == "number":
            L.append(round(random.random()*100))
            num_count += 1

        #generating random words from 1-10 alphabets and add to the list
        alphabets = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
        if Type == "word":
            n_words = int(random.random()*3) + 1
            word = ''
            for n in range(n_words):
                n_alphabets = int(random.random()*10) + 1
                s = ''
                s = s.join([alphabets[int(round(random.random()*25))] for i in range(n_alphabets)])
                if n != 0:
                    word = word + ' ' + s
                else:
                    word = word + s
            L.append(word)
            word_count += n_words
            
    #print(num_count)
    #print(word_count)

    return(L)
````

````python
### part a
def count_procedural (L):
    '''
    Input: List of numbers and words
    Output: Returns none, but prints out counts of numbers and words
    The function to get counts of numbers and words using procedural code
    '''
    no_count = 0
    word_count = 0

    for x in range(len(L)):
        if isinstance(L[x], int):
            no_count += 1 #no_count + 1
        elif isinstance(L[x], str):
            word_count = word_count + L[x].count(' ') + 1
            #word_count = word_count + 1
    
    print("Count of Numbers: ", no_count)
    print("Count of Words: ", word_count)
````

````python
### part b
def count_functional (L):

    '''
    Input: List of numbers and words
    Output: Returns none, but prints out counts of numbers and words
    The function to get counts of numbers and words using functional code
    '''

    numbers = [number for number in L if isinstance(number,int)]
    no_count = len(numbers)
    word_count = [string.count(' ') + 1 for string in L if isinstance(string, str)]
    word_count = sum(word_count)#len(L)-no_count
    
    print("Count of Numbers: ", no_count)
    print("Count of Words: ", word_count)
````

## Comparison

````python
import timeit
N = [100, 1000, 10000, 100000]

### create two separate lists to save count_type_loop and save count_type_lc run time.
loop = []
lc = []
for n in N:
  l = genElement(n)
  loop.append(timeit.timeit("lambda: count_procedural(l)", number=1))
  lc.append(timeit.timeit("lambda: count_functional(l)", number=1))

import matplotlib.pyplot as plt
x = N
k1 = loop
k2 = lc
plt.plot(x, k1, 's-', color = 'r', label = "Loop Method")
plt.plot(x, k2, 'o-', color = 'g', label = "List Comprehension Method")
plt.xlabel("N")
plt.ylabel("TimeIt result")
plt.xscale('log')
plt.legend(loc = "best")
plt.title("Comparison bewteen Loop and List Comprehension Method over different N")
plt.show()
````

## Sorted examples

* https://docs.python.org/3/library/functions.html#sorted

````python
#key parameter provides an ordering of the elements

key = str.lower

string_lst = ['This', 'is', 'a', 'test', 'string', 'from', 'Andrew']

#default: 'A' < 'a'

print(sorted(string_lst))

#when comparing lower case, 'a' > 'andrew', 'test' > 'this'

print(sorted(string_lst, key=key))

# when you compare A and B, you actually compare key(A) and key(B).

def bubble_sort(lst,key):
    for i in range(len(lst)):
        for j in range(i+1,len(lst)):
            if key(lst[i]) > key(lst[j]): #element comparison
                lst[i], lst[j] = lst[j], lst[i]
    return lst

lower_case_lst = bubble_sort(string_lst,key)

print(lower_case_lst)

def key(x):
    return x

default_lst = bubble_sort(string_lst,key)

print(default_lst)
````

````python
#key parameter also makes the elements comparable

test = [90, 'a', 'z', 'c']

def key_mixed(x):
    return str(x)

mixed_lst = bubble_sort(test,key_mixed)

print(mixed_lst)
print(sorted(test,key=key_mixed))

#numbers come before alphabets in strings 
````

````python
import random

class Dog:

    # class variables shared by all instances
    kind = 'canine'         

    # instance variables unique to each instance
    def __init__(self, name):
        self.name = name    
        self.tricks = []
    
    def change_kind(self, new_kind):
        self.kind = new_kind
    
    def add_trick(self, trick):
        self.tricks.append(trick)
    
    def learn_trick_from(self, trick):
        self.add_trick(trick)

Fido = Dog('Fido')
Buddy = Dog('Buddy')

Fido.add_trick('play dead')
Fido.add_trick('roll over')
Buddy.learn_trick_from(random.choice(Fido.tricks))

print(Buddy.tricks)
````
