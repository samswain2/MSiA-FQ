### Load functions

```python
# Load libs
import random
import string

# Create string of letters
letters = string.ascii_lowercase + string.ascii_uppercase

# Generate random list procedurally
def gen_list(n):
    # Get random list of 0 and 1
    lst = [random.randint(0,1) for _ in range(n)]
    
    # Generate list of numbers and words
    t = []
    for i in lst:
        if i == 0:
            t.append("".join(random.sample(letters, random.choice(range(1, 10)))))
        elif i == 1:
            t.append(random.choice(range(-1000, 1000)))
   
    # Combine words that're next to other words
    temp = []
    final_lst = []
    for i in t:
        if (isinstance(i, str)):
            temp.append(i)
        elif (isinstance(i, int)):
            if temp:
                final_lst.append(" ".join(temp))
                temp = []
            elif not temp:
                pass
            final_lst.append(i)

    if isinstance(t[-1], str):
        final_lst.append(" ".join(temp))

    return final_lst

# Check for spaces
def check_space(s):
    # Count
    count = 0
    # Check each index
    for i, v in enumerate(s):
        # Check each char
        # is blank or not
        if s[i] == " ":
            count += 1
         
    return count

# Get amount to add to word or number
def get_count(a):
    # Check if a is int or str
    if isinstance(a, int):
        return 1
    elif isinstance(a, str):
        return check_space(a) + 1

# Procedural method    
def get_n_and_w_for(lst):
    # Initalize number and word counts
    ns = 0
    ws = 0
    # Count numbers and words
    for i in lst:
        if isinstance(i, int):
            ns += get_count(i)
        elif isinstance(i, str):
            ws += get_count(i)
    
    return ns, ws

# Functional method
def get_n_and_w_lc(lst):
    # Count number of numbers
    ns = sum([get_count(i) for i in lst if isinstance(i, int)])
    # Count number of words
    ws = sum([get_count(i) for i in lst if isinstance(i, str)])
    
    return ns, ws

# Record time of functions
import timeit
weight = 1
def time_ns(n):
    x = gen_list(n)
    t_p = timeit.Timer(lambda: get_n_and_w_for(x))
    t_f = timeit.Timer(lambda: get_n_and_w_lc(x))
    p_vs_f = {"procedural": t_p.timeit(1) * weight, 
              "functional": t_f.timeit(1) * weight}
    
    return p_vs_f
```

### Calculate and plot results

```python
# Create dicts for all sizes of n
d_timeit = {}
for i in [100, 1000, 10000, 100000]:
    d_timeit[str(i)] = time_ns(i)

# Plot results
# Plotting libs
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure

%matplotlib inline

# Plot performance for each n
fig, ax = plt.subplots(2, 2)
fig.set_size_inches(18.5, 10.5)

fig.suptitle('Procedural vs Functional Programming Performance', fontsize = 26)
ax[0, 0].bar(d_timeit["100"].keys(), d_timeit["100"].values(), color='g')
ax[0, 0].set_title("N = 100")
ax[0, 1].bar(d_timeit["1000"].keys(), d_timeit["1000"].values(), color='g')
ax[0, 1].set_title("N = 1000")
ax[1, 0].bar(d_timeit["10000"].keys(), d_timeit["10000"].values(), color='g')
ax[1, 0].set_title("N = 10000")
ax[1, 1].bar(d_timeit["100000"].keys(), d_timeit["100000"].values(), color='g')
ax[1, 1].set_title("N = 100000")
plt.show()
```

```python
# Get data for line plots
time_pro = []
time_fun = []
for i in d_timeit:
    time_pro.append(d_timeit[i]["procedural"])
    time_fun.append(d_timeit[i]["functional"])


fig  = plt.figure()
fig.set_size_inches(18.5, 10.5)
plt.plot(time_pro, label = "procedural")
plt.plot(time_fun, label = "functional")
plt.title("Procedural vs Functional Programming Speed", fontsize = 26)
plt.legend(fontsize = 20)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.show()
```

### Create menu & menu functionality

```python
# Load csv lib
import csv

# Create menu text function
def print_menu():
    print("Grade System Menu:")
    print("1: Load Data (Do this first)")
    print("2: Print List of Student Names, Final Scores, and Letter Grade")
    print("3: Print Score Summary")
    print("4: Identify Values in Top 5%")
    print("5: Display Pie Chart of Final Letter Grades")
    print("6: Create Box Plot Parameters")
    print("7: Exit System")
    
# Load usable menu
def menu():
    # Pre-set vars
    rf = ["1", "2", "3", "4", "5", "6", "7"]
    print_menu()
    dl = False
    ui = None
    
    # Initiate while loop to use menu
    while ui != "7":
        print("-----------------------------")
        # Prompt user for input
        ui = input("Please select a task: ")
        
        # Menu option 1 (Load data) 
        if ui == "1":
            # Load in/create dict with data
            # 'id': ['Name', 'Quizzes', 'Projects', 'Exams', 'Final_Grade']
            df = {}
            with open('exams.csv', 'r') as file:
                my_reader = csv.reader(file, delimiter = ',')
                for row in my_reader:
                    df[row[0]] = row[1:]   

            del df['id']

            # Save column indices for easy access
            Name = 0
            Quizzes = 1
            Projects = 2
            Exams = 3
            Final_Grade = 4
            Letter_Grade = 5
            
            # Get final grades
            def get_fg(lst):
                return round(int(lst[Quizzes]) * .3 + 
                             int(lst[Projects]) * .3 + 
                             int(lst[Exams]) * .4, 2)
            for i in df:
                df[i].append(get_fg(df[i]))

            # Convert grades to ints
            for i in df:
                df[i][Quizzes] = int(df[i][Quizzes])
                df[i][Projects] = int(df[i][Projects])
                df[i][Exams] = int(df[i][Exams])

            # Get letter grades
            def get_lg(lst):
                if lst[Final_Grade] >= 90:
                    return "A"
                elif 80 <= lst[Final_Grade] < 90:
                    return "B"
                elif 20 <= lst[Final_Grade] < 80:
                    return "C"
                elif 10 <= lst[Final_Grade] < 20:
                    return "D"
                elif 0 <= lst[Final_Grade] < 10:
                    return "F"
                
            for i in df:
                df[i].append(get_lg(df[i]))
                
            # Get lists to calculate sample statistics
            import statistics
            import math
            t_ss = []
            for i in df:
                t_ss.append(df[i][Final_Grade])
            
            t_sss = []
            for i in t_ss:
                t_sss.append(math.floor(i))
            
            # Get list of letter grades
            lgs = []
            for i in df:
                lgs.append(df[i][Letter_Grade])
            
            # Print/store results
            print("Data Loaded")
            dl = True
            
        # Menu option 2 (List of student names, final scores, and letter grades sorted by names)
        elif ui == "2":
            # Check if data is loaded
            if dl == True:
                t_2 = []
                for i in df:
                    # Get student names, final score, and letter grade
                    t_2.append([df[i][x] for x in [Name, Final_Grade, Letter_Grade]])
                    t_2.sort(key = lambda x: x[0])
                # Print student names, final score, and letter grades
                for row in t_2:
                    print(row)
                    
            else:
                print("Please load data before performing other tasks")
        
        # Menu option 3 (score summary)
        elif ui == "3":
            # Check if data is loaded
            if dl == True:
                # Calculate and print sample statistics
                print("Summary Statistics:")
                print("---------------------")
                print(f"Student Count: {len(t_ss)}")
                print(f"Minimum Score: {min(t_ss)}")
                print(f"Maximum Score: {max(t_ss)}")
                print(f"Mean of Scores: {round(statistics.mean(t_ss), 2)}")
                print(f"Mode of Scores: {statistics.mode(t_sss)}")
                print(f"Mean of Scores: {round(statistics.stdev(t_ss), 2)}")
                
            else:
                print("Please load data before performing other tasks")  
        
        # Menu option 4 (Identify values that are larger than the mean and two times standard deviation)
        elif ui == "4":
            # Check if data is loaded
            if dl == True:
                # Calculate cutoff (mean + 2*sd)
                x_bar = statistics.mean(t_ss)
                x_sd = statistics.stdev(t_ss)
                cutoff_4 = x_bar + 2 * x_sd
                
                # Check all datapoints and print those above cutoff
                for i in df:
                    if df[i][Final_Grade] > cutoff_4:
                        print(f"{df[i][Name]}: {df[i][Final_Grade]}")
            else:
                print("Please load data before performing other tasks")       
        
        # Menu option 5 (Plot a pie chart showing the final letter grades distribution)
        elif ui == "5":
            # Check if data is loaded
            if dl == True:
                # Import plotting lib
                import matplotlib.pyplot as plt
                
                # Get count of each letter grade for plot
                df_lgs = {"A": 0, "B": 0, "C": 0, "D": 0, "F": 0}
                for i in df_lgs:
                    df_lgs[i] = lgs.count(i)

                # Plot data
                plt.pie(df_lgs.values(), labels = df_lgs.keys())
                plt.show()
                
            else:
                print("Please load data before performing other tasks")  
        
        # Menu option 6 (Create box plots parameters)
        elif ui == "6":
            # Check if data is loaded
            if dl == True:
                # Create temp sorted list to calculate Q1 and Q3
                t_ss_6 = sorted(t_ss)
                # Create and store boxplot parameters
                d_6 = {} 
                d_6["min"] = min(t_ss)
                d_6["max"] = max(t_ss)
                d_6["medial"] = statistics.median(t_ss)
                d_6["Q1"] = t_ss_6[int(100*0.25)]
                d_6["Q3"] = t_ss_6[int(100*0.75)]
                print(d_6)
                print("Boxplot parameters created and can be seen above")
                
            else:
                print("Please load data before performing other tasks")  
        
        # Menu option 7 (Exit menu)
        elif ui == "7":
            # Exit menu/message functionality
            import time
            time.sleep(.75)
            print("Thank you for using my app")
            time.sleep(.75)
            print(".")
            time.sleep(.75)
            print(".")
            time.sleep(.75)
            print(".")
            time.sleep(1)
            print("Goodbye")
            break
                
        # Inform user to put input a number from the menu        
        else:
            print("Error: Please pick an option from the menu above")
            
        # Print menu again for ease of use
        print("-----------------------------")
        print(f"|Task {ui} output above        |")
        print("-----------------------------")
        print("|Please select another task |")
        print("-----------------------------")
        print_menu()
menu()
```
